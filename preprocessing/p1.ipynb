```python
import os
import glob
import datetime
import pandas as pd
```

```python
dir = r'../../datasets/weekly'

def list_files(dir):
    csv_files = glob.glob(os.path.join(dir, "*.csv"))
    return csv_files
```

```python
def get_dataframes(csv_files):
    df = pd.DataFrame()
    for f in csv_files:
        tmp = pd.read_csv(f)
        df = df.append(tmp)

    return df
```

```python
csv_files = list_files(dir)
df = get_dataframes(csv_files)
df.head()
```

```python
df.reset_index(drop=True, inplace=True)
df
```

```python
row = list(df)
size = len(row)

df_sum = df.loc[:, row[3]:row[size-3]].sum(axis=1).reset_index()
df_sum.rename(columns={0:'sum'}, inplace=True)
df_sum
```

```python
df_sum["target"] = df["target"]
df_sum["target_cat"] = df["target_cat"]
df_sum
```

```python
today_ts = datetime.datetime.now()
today_ts = today_ts.strftime("%d-%m-%Y-%H-%M")

df_sum.to_csv("../preprocessing-results/p1.csv".format(today_ts), sep=',', encoding='utf-8', index=False)
```
